```python
# no 1 :Ada berapa rumah yang mengizinkan membawa hewan peliharaan dengan harga sewa per tahun di bawah 5000
print("Banyak yang mengizinkan hewan peliharaan : ", len(df.loc[df["animal"] == "acept"]))
```

```python
# no 2 : Data rumah dengan jumlah kamar mandi terbanyak
df.loc[df["bathroom"] == df["bathroom"].max()]
```

```python
# no 3 : Ada berapa rumah dengan luas di atas 200 dan pajak bangunan per tahun di bawah 1700?.
print("jumlah rumah dengan luas di atas 200 dan pajak bangunan di bawah 1700: ",len(df.loc[df["area"] > 200].loc[df["property_tax"] < 1700]))
```

```python
# no 4 : Banyak rumah dengan biaya asuransi kebakaran per tahun di atas 70 untuk masing-masing kota. 
df.loc[df["fire_insurance"] > 70]["city"].value_counts().rename_axis("Kota").reset_index(name="Jumlah rumah dengan asuransi kebakaran di atas 70")
```

```python
# no 5 : 5 data rumah dengan jumlah lantai terbanyak yang belum dilengkapi furnitur (not furnished).
df.loc[df["furniture"] == "not furnished"].sort_values(by="floor", ascending=False).head(5).reset_index(drop=True)
```
