```python
# no 1 : Banyaknya rumah masing-masing tipe kamar
df["room_type"].value_counts().rename_axis("Tipe_kamar").reset_index(name="Jumlah mahasiswa")
```

```python
# no 2 : Jumlah review terbanyak
print("Jumlah reviwe terbanyak: ",df["number_of_reviews"].max())
```

```python
# no 3 : 5 data dengan harga per malam termahal
df.sort_values(by="price", ascending=False).head(5).reset_index(drop=True)
```

```python
# no 4 : Rumah dengan pemilik bernama Richard
df.loc[df["host_name"] == "Richard"].reset_index(drop=True)
```

```python
# no 5 : Koefisien korelasi antara harga rumah per malam dengan jumlah malam penyewaan minimum. Apakah
# berkorelasi lurus, terbalik, atau tidak berkolerasi? Jelaskan 
df["price"].corr(df["minimum_nights"])
```

```python
# maka dari angka tersebut data tidak berkolerasi karena nilainnya hanya 0.01 yang dapat dilihat juga dari 
# grafik di bawah ini
import matplotlib.pyplot as plot
df.plot(kind="scatter", x="price", y="minimum_nights", xlabel="harga per malam",ylabel="jumlah malam minimum", label="Korelasi harga per malam dengan jumlah malam minimum")
plot.show()
```
