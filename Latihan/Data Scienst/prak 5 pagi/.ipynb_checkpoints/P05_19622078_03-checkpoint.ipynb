```python
# no 2 : Pie chart perbandingan banyaknya masing-masing jenjang pendidikan terakhir dari pengguna dengan banyak transaksi selama 12 bulan terakhir di atas 30.
df.loc[df["Total_Trans_Amt"] > 30]["Education_Level"].value_counts().plot(kind="pie")
```
