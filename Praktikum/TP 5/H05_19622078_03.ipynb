```python
df["nilai_mat"].plot(kind="hist", rwidth=0.8, title="Distribusi Nilai Matematika")
plt.show()
```

```python
df["fakultas"].value_counts().plot(kind="bar", title="Banyak Mahasiswa ITB", legend="best", label="Jumlah Mahasiswa")
plt.show()
```

```python
df["fakultas"].value_counts().plot(kind="pie", title="Banyak Mahasiswa ITB",label="")
plt.show()
```

```python
print("Fakultas dengan mahasiswa paling banyak: ",df["fakultas"].value_counts().idxmax())
```
